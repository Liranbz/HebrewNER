```python
data_path = 'resources' + os.sep + 'X_test_with_predictions.csv'    
preds_df = pd.read_csv(data_path)
```

```python
from sklearn.metrics import classification_report
```

```python
classes_without_O = ['U-PERCENT', 'L-PERS', 'U-PERS', 'L-ORG', 'L-LOC', 'I-ORG', 'I-LOC', 'B-ORG', 'L-DATE', 'I-MONEY', 'B-MISC', 'L-MISC', 'L-MONEY', 'B-LOC', 'B-PERS', 'I-PERS', 'U-DATE', 'B-DATE', 'U-LOC', 'B-MONEY', 'U-MISC', 'I-MISC', 'I-DATE', 'L-PERCENT', 'I-TIME', 'U-ORG', 'L-TIME', 'B-PERCENT', 'B-TIME', 'U-TIME', 'I-PERCENT', 'U-MONEY' ]
```

```python
y_pred = list(preds_df['y_pred'].values)
y_true = list(preds_df['y_test'].values)
```

```python
y_true = pd.Series(y_true)
y_pred = pd.Series(y_pred)
cross_tab = pd.crosstab(y_true, y_pred, rownames=['Real Label'], colnames=['Prediction'], margins=True)
report = classification_report(y_true, y_pred, labels=classes_without_O, target_names=classes_without_O)
report_with_O = classification_report(y_true, y_pred)
```

```python
print(report)
```

```python
transformed_y_true = []
for label in y_true:
    if len(label.split("-"))>1:
        label = label.split("-")[1]
    transformed_y_true.append(label)
```

```python
transformed_y_pred = []
for label in y_pred:
    if len(label.split("-"))>1:
        label = label.split("-")[1]
    transformed_y_pred.append(label)
```

```python
tags_without_O = list(set(list(y_true)))
tags_without_O.remove('O')
y_true = pd.Series(transformed_y_true)
y_pred = pd.Series(transformed_y_pred)
cross_tab = pd.crosstab(y_true, y_pred, rownames=['Real Label'], colnames=['Prediction'], margins=True)
report = classification_report(y_true, y_pred, labels=tags_without_O, target_names=tags_without_O)
report_with_O = classification_report(y_true, y_pred)
```

```python
print(report)
```
